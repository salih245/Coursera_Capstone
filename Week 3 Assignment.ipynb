```python
# importing libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
from IPython.display import display_html
import numpy as np
from geopy.geocoders import Nominatim 
from pandas.io.json import json_normalize  

from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

!pip -q install folium
import folium 

from IPython.display import Image
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
df_info = pd.read_html(tab)
df=df_info[0]
df.head()
```

```python
# Deleting the rows where Borough is 'Not assigned'
df_without_Not_assigned = df[df.Borough != 'Not assigned']

# Combining the neighbourhood within Postal Code
df_combinging = df_without_Not_assigned.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_combinging.reset_index(inplace=True)

# Replacing the name of the Neighbourhood
df_combinging['Neighbourhood'] = np.where(df_combinging['Neighbourhood'] == 'Not assigned',df_combinging['Borough'], df_combinging['Neighbourhood'])

#Display row 11
df_combinging.head(12)
```

```python
#print the number of rows of your dataframe
df_combinging.shape
```

```python
#Use the Geocoder package or the csv file to create the following dataframe:
df_lon = pd.read_csv('https://cocl.us/Geospatial_data')
df_lon.head()
```

```python
df_map = pd.merge(df_combinging,df_lon, on='Postal Code')
df_map.head(12)
```

```python
#the data frame which contains Toronto
df_toronto = df_map[df_map['Borough'].str.contains('Toronto',regex=False)]
df_toronto
```

```python
#Visualizing all the Neighbourhoods
toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for latitude,longitude,borough,neighbourhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [latitude,longitude],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(toronto_map)
toronto_map
```

```python
k=5
toronto_clustering = df_toronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create cluster_toronto_map
cluster_toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for latitude, longitude, neighbourhood, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighbourhood'], df_toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [latitude,longitude],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(cluster_toronto_map)
       
cluster_toronto_map
```
